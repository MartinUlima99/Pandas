```python
import pandas as pd

# crear dataframe de ejemplo
data = {'Nombre': ['Bulbasaur', 'Charizard', 'Pikachu', 'Squirtle', 'Snorlax'],
        'Tipo': ['Grass', 'Fire', 'Electric', 'Water', 'Normal'],
        'Ataque': [49, 84, 55, 48, 110]}

pokemon_df = pd.DataFrame(data)

# filtrar dataframe usando loc y dos condiciones
poke_filtrado = pokemon_df.loc[(pokemon_df['Tipo'] == 'Grass') & (pokemon_df['Ataque'] > 90)]

# imprimir resultado
print(poke_filtrado)
```
