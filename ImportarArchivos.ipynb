**Agrupación de datos**

```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {'Nombre': ['Ana', 'Pedro', 'María', 'Juan', 'Sara'],
        'Edad': [25, 30, 28, 32, 27],
        'Género': ['Femenino', 'Masculino', 'Femenino', 'Masculino', 'Femenino'],
        'Ciudad': ['Madrid', 'Barcelona', 'Madrid', 'Madrid', 'Barcelona']}
df = pd.DataFrame(data)

### Agrupar los datos por ciudad y contar el número de personas en cada ciudad
df_grouped = df.groupby('Ciudad').size()
print(df_grouped)
```

```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {'Nombre': ['Ana', 'Pedro', 'María', 'Juan', 'Sara'],
        'Edad': [25, 30, 28, 32, 27],
        'Género': ['Femenino', 'Masculino', 'Femenino', 'Masculino', 'Femenino'],
        'Ciudad': ['Madrid', 'Barcelona', 'Madrid', 'Madrid', 'Barcelona'],
        'Profesión': ['Ingeniera', 'Doctor', 'Abogada', 'Ingeniero', 'Abogada'],
        'Salario': [3000, 5000, 4000, 3500, 4500],
        'Años_experiencia': [3, 5, 4, 7, 6]}
df = pd.DataFrame(data)

### Agrupar los datos por género y obtener la media del salario y la suma de los años de experiencia por grupo
df_grouped = df.groupby('Género').agg({'Salario': 'mean', 'Años_experiencia': 'sum'})
print(df_grouped)
```

**Pivot Table**

```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {'Ciudad': ['A', 'A', 'B', 'B', 'B'],
        'Año': [2018, 2019, 2018, 2019, 2020],
        'Ventas': [100, 150, 200, 250, 300]}
df = pd.DataFrame(data)

# Crear la tabla pivote básica
pivot_table_basic = df.pivot_table(index='Ciudad', columns='Año', values='Ventas')
df
#print(pivot_table_basic)
```

```python
print(pivot_table_basic)
```

```python
# Crear una tabla pivote con agregación de funciones
pivot_table_intermediate = df.pivot_table(index='Ciudad', columns='Año',
                                          values='Ventas', aggfunc=[sum, len])
print(pivot_table_intermediate)
```

```python
# Crear una tabla pivote con múltiples índices y columnas
pivot_table_advanced = df.pivot_table(index=['Ciudad', 'Año'], columns='Ventas',
                                      values='Ventas', aggfunc=len, fill_value=0)
print(pivot_table_advanced)
```
